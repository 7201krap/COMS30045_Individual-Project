```python
import matplotlib.pyplot as plt
import numpy as np
import re
import ast 
import torch
```

```python
def val_extractor(file_name1, file_name2, file_name3):    
    # seed 1
    seed1_spar12, seed1_sele12 = 0, 0
    seed1_spar23, seed1_sele23 = 0, 0
    seed1_spar34, seed1_sele34 = 0, 0
    seed1_spar45, seed1_sele45 = 0, 0
    
    # seed 100
    seed2_spar12, seed2_sele12 = 0, 0
    seed2_spar23, seed2_sele23 = 0, 0
    seed2_spar34, seed2_sele34 = 0, 0
    seed2_spar45, seed2_sele45 = 0, 0
    
    # seed 1234
    seed3_spar12, seed3_sele12 = 0, 0
    seed3_spar23, seed3_sele23 = 0, 0
    seed3_spar34, seed3_sele34 = 0, 0
    seed3_spar45, seed3_sele45 = 0, 0
    
#     index 2 --> spar12 
#     index 3 --> spar23 
#     index 4 --> spar34 
#     index 5 --> spar45 
    
#     index 8 --> selec12 
#     index 10 --> selec23 
#     index 12 --> selec34 
#     index 14 --> selec45
    
    # seed 1 
    with open(file_name1) as fi:
        for i, line in enumerate(fi):   
            if i == 2:
                seed1_spar12 = ast.literal_eval(line)
            if i == 3:
                seed1_spar23 = ast.literal_eval(line)
            if i == 4:
                seed1_spar34 = ast.literal_eval(line)
            if i == 5:
                seed1_spar45 = ast.literal_eval(line)
            if i == 8:
                seed1_sele12 = ast.literal_eval(line) 
            if i == 10: 
                seed1_sele23 = ast.literal_eval(line)
            if i == 12:
                seed1_sele34 = ast.literal_eval(line)
            if i == 14:
                seed1_sele45 = ast.literal_eval(line)
    
    # seed 100 
    with open(file_name2) as fi:
        for i, line in enumerate(fi):   
            if i == 2:
                seed2_spar12 = ast.literal_eval(line)
            if i == 3:
                seed2_spar23 = ast.literal_eval(line)
            if i == 4:
                seed2_spar34 = ast.literal_eval(line)
            if i == 5:
                seed2_spar45 = ast.literal_eval(line)
            if i == 8:
                seed2_sele12 = ast.literal_eval(line) 
            if i == 10: 
                seed2_sele23 = ast.literal_eval(line)
            if i == 12:
                seed2_sele34 = ast.literal_eval(line)
            if i == 14:
                seed2_sele45 = ast.literal_eval(line)
    
    # seed 100           
    with open(file_name3) as fi:
        for i, line in enumerate(fi):   
            if i == 2:
                seed3_spar12 = ast.literal_eval(line)
            if i == 3:
                seed3_spar23 = ast.literal_eval(line)
            if i == 4:
                seed3_spar34 = ast.literal_eval(line)
            if i == 5:
                seed3_spar45 = ast.literal_eval(line)
            if i == 8:
                seed3_sele12 = ast.literal_eval(line) 
            if i == 10: 
                seed3_sele23 = ast.literal_eval(line)
            if i == 12:
                seed3_sele34 = ast.literal_eval(line)
            if i == 14:
                seed3_sele45 = ast.literal_eval(line)

    spar12, spar12_se = list(), list()
    sele12, sele12_se = list(), list()
    
    spar23, spar23_se = list(), list()
    sele23, sele23_se = list(), list()
    
    spar34, spar34_se = list(), list()
    sele34, sele34_se = list(), list()
    
    spar45, spar45_se = list(), list()
    sele45, sele45_se = list(), list()

    for i in range(len(seed1_spar12)):
        spar12.append   (np.mean([seed1_spar12[i], seed2_spar12[i], seed3_spar12[i]]))
        spar12_se.append(np.std ([seed1_spar12[i], seed2_spar12[i], seed3_spar12[i]]) / np.sqrt(3))
        sele12.append   (np.mean([seed1_sele12[i], seed2_sele12[i], seed3_sele12[i]]))
        sele12_se.append(np.std ([seed1_sele12[i], seed2_sele12[i], seed3_sele12[i]]) / np.sqrt(3))
        
        spar23.append   (np.mean([seed1_spar23[i], seed2_spar23[i], seed3_spar23[i]]))
        spar23_se.append(np.std ([seed1_spar23[i], seed2_spar23[i], seed3_spar23[i]]) / np.sqrt(3))
        sele23.append   (np.mean([seed1_sele23[i], seed2_sele23[i], seed3_sele23[i]]))
        sele23_se.append(np.std ([seed1_sele23[i], seed2_sele23[i], seed3_sele23[i]]) / np.sqrt(3))
        
        spar34.append   (np.mean([seed1_spar34[i], seed2_spar34[i], seed3_spar34[i]]))
        spar34_se.append(np.std ([seed1_spar34[i], seed2_spar34[i], seed3_spar34[i]]) / np.sqrt(3))
        sele34.append   (np.mean([seed1_sele34[i], seed2_sele34[i], seed3_sele34[i]]))
        sele34_se.append(np.std ([seed1_sele34[i], seed2_sele34[i], seed3_sele34[i]]) / np.sqrt(3))
        
        spar45.append   (np.mean([seed1_spar45[i], seed2_spar45[i], seed3_spar45[i]]))
        spar45_se.append(np.std ([seed1_spar45[i], seed2_spar45[i], seed3_spar45[i]]) / np.sqrt(3))
        sele45.append   (np.mean([seed1_sele45[i], seed2_sele45[i], seed3_sele45[i]]))
        sele45_se.append(np.std ([seed1_sele45[i], seed2_sele45[i], seed3_sele45[i]]) / np.sqrt(3))
      
    return spar12, spar12_se, spar23, spar23_se, spar34, spar34_se, spar45, spar45_se, sele12, sele12_se, sele23, sele23_se, sele34, sele34_se, sele45, sele45_se 
```

```python
def test_val_extractor(file_name1, file_name2, file_name3):    
    # seed 1
    seed1_acc = 0
    
    # seed 100
    seed2_acc = 0
    
    # seed 1234
    seed3_acc = 0

    # index 0 --> test acc
    
    # seed 1 
    with open(file_name1) as fi:
        for i, line in enumerate(fi):   
            if i == 0:
                seed1_acc = ast.literal_eval(line)
    
    # seed 1 
    with open(file_name2) as fi:
        for i, line in enumerate(fi):   
            if i == 0:
                seed2_acc = ast.literal_eval(line)
    
    # seed 1 
    with open(file_name3) as fi:
        for i, line in enumerate(fi):   
            if i == 0:
                seed3_acc = ast.literal_eval(line)

    acc, acc_se = list(), list()
    
    for i in range(len(seed1_acc)):
        acc.append   (np.mean([seed1_acc[i], seed2_acc[i], seed3_acc[i]]))
        acc_se.append(np.std ([seed1_acc[i], seed2_acc[i], seed3_acc[i]]) / np.sqrt(3))
        
    return acc, acc_se
```
