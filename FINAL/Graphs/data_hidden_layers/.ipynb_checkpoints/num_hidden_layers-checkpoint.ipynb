```python
import matplotlib.pyplot as plt
import numpy as np
import re
import ast 
import torch
```

```python
def test_val_extractor(file_name):    
    # get accuracy from sparsity.txt
    test_val_accumulator = list()
    with open(file_name) as fi:
        for i, line in enumerate(fi):
            if i == 0:
                line = ast.literal_eval(line)
                test_val_accumulator = line
                
    # return list of test accuracy 
    return test_val_accumulator
```

```python
def selectivity_val_extractor(file_name):
    selectivity_average = 0
    selectivity_std = 0
    with open(file_name) as fi:
        for i, line in enumerate(fi):
            if i == 1:
                selectivity_average = ast.literal_eval(line)
            if i == 2:
                selectivity_std = ast.literal_eval(line)
    return selectivity_average, selectivity_std
```

```python
def sparsity_val_extractor(file_name):
    sparsity_average = 0
    with open(file_name) as fi:
        for i, line in enumerate(fi):
            if i == 1:
                sparsity_average = ast.literal_eval(line)
    return sparsity_average
```

```python
def sparsity_all_plotter(HL1, HL2, HL3, HL4, HL5, location='best'):
    fig, ax = plt.subplots(figsize=(20, 15))
    
    ax.plot(np.linspace(0.0, 50.0, num=50), HL1, label='1 Hidden layer')
    ax.plot(np.linspace(0.0, 50.0, num=50), HL2, label='2 Hidden layers')
    ax.plot(np.linspace(0.0, 50.0, num=50), HL3, label='3 Hidden layers')
    ax.plot(np.linspace(0.0, 50.0, num=50), HL4, label='4 Hidden layers')
    ax.plot(np.linspace(0.0, 50.0, num=50), HL5, label='5 Hidden layers')
    ax.set_title('Sparsity')
    ax.set_ylim([-0.2,1.2])
    ax.legend(loc=location)
    plt.show()
```

```python
def selectivity_all_plotter(HL1, HL2, HL3, HL4, HL5, location='best'):
    fig, ax = plt.subplots(figsize=(20, 15))
    
    ax.plot(np.linspace(0.0, 50.0, num=50), HL1, label='1 Hidden layer')
    ax.plot(np.linspace(0.0, 50.0, num=50), HL2, label='2 Hidden layers')
    ax.plot(np.linspace(0.0, 50.0, num=50), HL3, label='3 Hidden layers')
    ax.plot(np.linspace(0.0, 50.0, num=50), HL4, label='4 Hidden layers')
    ax.plot(np.linspace(0.0, 50.0, num=50), HL5, label='5 Hidden layers')
    ax.set_title('Selectivity')
    ax.set_ylim([-0.2,1.2])
    ax.legend(loc=location)
    plt.show()
```

```python
def testacc_all_plotter(HL1, HL2, HL3, HL4, HL5, location='best'):
    fig, ax = plt.subplots(figsize=(20, 15))
    
    ax.plot(np.linspace(0.0, 50.0, num=50), HL1, label='1 Hidden layer')
    ax.plot(np.linspace(0.0, 50.0, num=50), HL2, label='2 Hidden layers')
    ax.plot(np.linspace(0.0, 50.0, num=50), HL3, label='3 Hidden layers')
    ax.plot(np.linspace(0.0, 50.0, num=50), HL4, label='4 Hidden layers')
    ax.plot(np.linspace(0.0, 50.0, num=50), HL5, label='5 Hidden layers')
    ax.set_title('Test accuracy')
    ax.set_ylim([-0.2,1.2])
    ax.legend(loc=location)
    plt.show()
```

# Adadelta sparsity

```python
sparsity_avg_HL1 = sparsity_val_extractor('1HL_sparsity_Adadelta.txt')
sparsity_avg_HL2 = sparsity_val_extractor('2HL_sparsity_Adadelta.txt')
sparsity_avg_HL3 = sparsity_val_extractor('3HL_sparsity_Adadelta.txt')
sparsity_avg_HL4 = sparsity_val_extractor('4HL_sparsity_Adadelta.txt')
sparsity_avg_HL5 = sparsity_val_extractor('5HL_sparsity_Adadelta.txt')
sparsity_all_plotter(sparsity_avg_HL1, sparsity_avg_HL2, sparsity_avg_HL3, sparsity_avg_HL4, sparsity_avg_HL5)
```

```python
selectivity_avg_HL1, selectivity_std_HL1 = selectivity_val_extractor('1HL_selectivity_Adadelta.txt')
selectivity_avg_HL2, selectivity_std_HL2 = selectivity_val_extractor('2HL_selectivity_Adadelta.txt')
selectivity_avg_HL3, selectivity_std_HL3 = selectivity_val_extractor('3HL_selectivity_Adadelta.txt')
selectivity_avg_HL4, selectivity_std_HL4 = selectivity_val_extractor('4HL_selectivity_Adadelta.txt')
selectivity_avg_HL5, selectivity_std_HL5 = selectivity_val_extractor('5HL_selectivity_Adadelta.txt')
selectivity_all_plotter(selectivity_avg_HL1, selectivity_avg_HL2, selectivity_avg_HL3, selectivity_avg_HL4, selectivity_avg_HL5)
```

# Adadelta test accuracy

```python
test_HL1 = test_val_extractor('1HL_sparsity_Adadelta.txt')
test_HL2 = test_val_extractor('2HL_sparsity_Adadelta.txt')
test_HL3 = test_val_extractor('3HL_sparsity_Adadelta.txt')
test_HL4 = test_val_extractor('4HL_sparsity_Adadelta.txt')
test_HL5 = test_val_extractor('5HL_sparsity_Adadelta.txt')
testacc_all_plotter(test_HL1, test_HL2, test_HL3, test_HL4, test_HL5)
```
