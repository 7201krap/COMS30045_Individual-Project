```python
import matplotlib.pyplot as plt
import numpy as np
import re
import ast 
import torch
```

```python
def test_val_extractor(file_name):    
    # get accuracy from sparsity.txt
    test_val_accumulator = list()
    with open(file_name) as fi:
        for i, line in enumerate(fi):
            if i == 0:
                line = ast.literal_eval(line)
                test_val_accumulator = line
                
    # return list of test accuracy 
    return test_val_accumulator
```

```python
def selectivity_val_extractor(file_name):
    selectivity_average = 0
    selectivity_std = 0
    with open(file_name) as fi:
        for i, line in enumerate(fi):
            if i == 1:
                selectivity_average = ast.literal_eval(line)
            if i == 2:
                selectivity_std = ast.literal_eval(line)
    return selectivity_average, selectivity_std
```

```python
def sparsity_val_extractor(file_name):
    sparsity_average = 0
    with open(file_name) as fi:
        for i, line in enumerate(fi):
            if i == 1:
                sparsity_average = ast.literal_eval(line)
    return sparsity_average
```

```python
def all_plotter_test(test_adadelta, test_adagrad, test_sgd, test_adam, location='best'):
    fig, ax = plt.subplots(figsize=(10, 8))
    
    ax.plot(np.linspace(0.0, 30, num=30), test_adadelta, label='test_adadelta', c='red')
    ax.plot(np.linspace(0.0, 30, num=30), test_adagrad, label='test_adagrad', c='blue')
    ax.plot(np.linspace(0.0, 30, num=30), test_sgd, label='test_sgd', c='green')
    ax.plot(np.linspace(0.0, 30, num=30), test_adam, label='test_adam', c='violet')
    
    ax.set_ylim([-0.2,1.2])
    ax.legend(loc=location)
    plt.show()
```

```python
def all_plotter_spar_selec(spar_adadelta, selec_adadelta, spar_adagrad, selec_adagrad, spar_sgd, selec_sgd, spar_adam, selec_adam, location='best'):
    fig, ax = plt.subplots(figsize=(10, 8))
    
    ax.plot(np.linspace(0.0, 30, num=30), spar_adadelta, label='spar_adadelta', c='red')
    ax.plot(np.linspace(0.0, 30, num=30), selec_adadelta, label='selec_adadelta', c='darkred')
    
    ax.plot(np.linspace(0.0, 30, num=30), spar_adagrad, label='spar_adagrad', c='blue')
    ax.plot(np.linspace(0.0, 30, num=30), selec_adagrad, label='selec_adagrad', c='darkblue')
    
    ax.plot(np.linspace(0.0, 30, num=30), spar_sgd, label='spar_sgd', c='green')
    ax.plot(np.linspace(0.0, 30, num=30), selec_sgd, label='selec_sgd', c='darkgreen')
    
    ax.plot(np.linspace(0.0, 30, num=30), spar_adam, label='spar_adam', c='violet')
    ax.plot(np.linspace(0.0, 30, num=30), selec_adam, label='selec_adam', c='darkviolet')
    
    ax.set_ylim([-0.2,1.2])
    ax.legend(loc=location)
    plt.show()
```

# Test accuracy

```python
test_acc_adadelta = test_val_extractor("sparsity_4_optim_Adadelta.txt")
test_acc_adagrad = test_val_extractor("sparsity_4_optim_Adagrad.txt")
test_acc_sgd = test_val_extractor("sparsity_4_optim_SGD.txt")
test_acc_adam = test_val_extractor("sparsity_4_optim_Adam.txt")

all_plotter_test(test_acc_adadelta, test_acc_adagrad, test_acc_sgd, test_acc_adam)
```

# Sparsity & Selectivity

```python
sparsity_adadelta = sparsity_val_extractor("sparsity_4_optim_Adadelta.txt")
sparsity_adagrad = sparsity_val_extractor("sparsity_4_optim_Adagrad.txt")
sparsity_sgd = sparsity_val_extractor("sparsity_4_optim_SGD.txt")
sparsity_adam = sparsity_val_extractor("sparsity_4_optim_Adam.txt")

selectivity_avg_adadelta, selectivity_std_adadelta = selectivity_val_extractor("selectivity_4_optim_Adadelta.txt")
selectivity_avg_adagrad, selectivity_std_adagrad = selectivity_val_extractor("selectivity_4_optim_Adagrad.txt")
selectivity_avg_sgd, selectivity_std_sgd = selectivity_val_extractor("selectivity_4_optim_SGD.txt")
selectivity_avg_adam, selectivity_std_adam = selectivity_val_extractor("selectivity_4_optim_Adam.txt")

all_plotter_spar_selec(sparsity_adadelta, selectivity_avg_adadelta, sparsity_adagrad, selectivity_avg_adagrad, sparsity_sgd, selectivity_avg_sgd, sparsity_adam, selectivity_avg_adam)
```
