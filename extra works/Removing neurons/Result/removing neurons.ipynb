```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
import re
import ast 
import torch
```

```python
def val_extractor(file_name1):    
    with open(file_name1) as fi:
        for i, line in enumerate(fi):
            if i == 0:
                accuracy = ast.literal_eval(line)   
    
    accuracy = accuracy[-1]
            
    return accuracy
```

```python
delete_cases = [1, 5, 10, 100]
optimizers = ['SGD', 'Adagrad', 'Adadelta', 'Adam']
seed_numbers = ['1234', '4321', '5678', '8765', '9999']
```

## Bring all data and Pre-process

```python
def get_deletion_cases(optimizer_name, sn):
    for dc in delete_cases:
        del1   = val_extractor(f'data/seed{sn}/{optimizer_name}/seed{sn}_neuronsize256_{optimizer_name}_delete1.txt')
        del5   = val_extractor(f'data/seed{sn}/{optimizer_name}/seed{sn}_neuronsize256_{optimizer_name}_delete5.txt')
        del10  = val_extractor(f'data/seed{sn}/{optimizer_name}/seed{sn}_neuronsize256_{optimizer_name}_delete10.txt')
        del100 = val_extractor(f'data/seed{sn}/{optimizer_name}/seed{sn}_neuronsize256_{optimizer_name}_delete100.txt')
        
    return [del1, del5, del10, del100]
```

```python
def mean_and_se(acc_seed1, acc_seed2, acc_seed3, acc_seed4, acc_seed5):
    
    acc_avg = list()
    acc_se  = list()

    for i in range(len(acc_seed1)):
        acc_avg.append(np.mean([acc_seed1[i], acc_seed2[i], acc_seed3[i], acc_seed4[i], acc_seed5[i]]))
        acc_se .append(np.std ([acc_seed1[i], acc_seed2[i], acc_seed3[i], acc_seed4[i], acc_seed5[i]]) / np.sqrt(5))
        
    return acc_avg, acc_se
```

```python
sgd1234 = get_deletion_cases('SGD', '1234')
sgd4321 = get_deletion_cases('SGD', '4321')
sgd5678 = get_deletion_cases('SGD', '5678')
sgd8765 = get_deletion_cases('SGD', '8765')
sgd9999 = get_deletion_cases('SGD', '9999')

grad1234 = get_deletion_cases('Adagrad', '1234')
grad4321 = get_deletion_cases('Adagrad', '4321')
grad5678 = get_deletion_cases('Adagrad', '5678')
grad8765 = get_deletion_cases('Adagrad', '8765')
grad9999 = get_deletion_cases('Adagrad', '9999')

delta1234 = get_deletion_cases('Adadelta', '1234')
delta4321 = get_deletion_cases('Adadelta', '4321')
delta5678 = get_deletion_cases('Adadelta', '5678')
delta8765 = get_deletion_cases('Adadelta', '8765')
delta9999 = get_deletion_cases('Adadelta', '9999')

adam1234 = get_deletion_cases('Adam', '1234')
adam4321 = get_deletion_cases('Adam', '4321')
adam5678 = get_deletion_cases('Adam', '5678')
adam8765 = get_deletion_cases('Adam', '8765')
adam9999 = get_deletion_cases('Adam', '9999')
```

```python
sgd_acc_avg, sgd_acc_se     = mean_and_se(sgd1234, sgd4321, sgd5678, sgd8765, sgd9999)
grad_acc_avg, grad_acc_se   = mean_and_se(grad1234, grad4321, grad5678, grad8765, grad9999)
delta_acc_avg, delta_acc_se = mean_and_se(delta1234, delta4321, delta5678, delta8765, delta9999)
adam_acc_avg, adam_acc_se   = mean_and_se(adam1234, adam4321, adam5678, adam8765, adam9999)
```

```python
length = np.linspace(1, 4, 4)

fig, ax = plt.subplots(figsize=(8, 8))

# default setting for grpahs 
ticks = ['1', '5', '10', '100']
ax.set_xticks(np.linspace(1, 4, 4))
ax.set_xticklabels(ticks)

# Accuracy 
# sgd
ax.plot(length, sgd_acc_avg, c='darkred', label='SGD')
ax.fill_between(length, np.subtract(sgd_acc_avg, sgd_acc_se), np.add(sgd_acc_avg, sgd_acc_se), color='lightcoral')

# adagrad
ax.plot(length, grad_acc_avg, c='darkgreen', label='Adagrad')
ax.fill_between(length, np.subtract(grad_acc_avg, grad_acc_se), np.add(grad_acc_avg, grad_acc_se), color='lime')

# adadelta
ax.plot(length, delta_acc_avg, c='darkblue', label='Adadelta')
ax.fill_between(length, np.subtract(delta_acc_avg, delta_acc_se), np.add(delta_acc_avg, delta_acc_se), color='cornflowerblue')

# adam
ax.plot(length, adam_acc_avg, c='purple', label='Adam')
ax.fill_between(length, np.subtract(adam_acc_avg, adam_acc_se), np.add(adam_acc_avg, adam_acc_se), color='violet')

ax.set_title('Test accuracy when varying the number of deleted neurons')
ax.set_xlabel('Deleted Neurons')
ax.set_ylabel('Test Accuracy')
ax.grid()

ax.legend()
```
