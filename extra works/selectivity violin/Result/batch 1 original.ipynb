```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
import re
import ast 
import torch
```

```python
def val_extractor(file_name1):    
    with open(file_name1) as fi:
        for i, line in enumerate(fi):
            if i == 4:
                each_selec = ast.literal_eval(line)     
            
    return each_selec
```

# Batch size = 1

## Adadelta

```python
delta64  = val_extractor('batch 1/seed1_batch1_neuronsize64_Adadelta.txt')
delta128 = val_extractor('batch 1/seed1_batch1_neuronsize128_Adadelta.txt')
delta256 = val_extractor('batch 1/seed1_batch1_neuronsize256_Adadelta.txt')
delta384 = val_extractor('batch 1/seed1_batch1_neuronsize384_Adadelta.txt')
delta512 = val_extractor('batch 1/seed1_batch1_neuronsize512_Adadelta.txt')
delta640 = val_extractor('batch 1/seed1_batch1_neuronsize640_Adadelta.txt')
delta768 = val_extractor('batch 1/seed1_batch1_neuronsize768_Adadelta.txt')

del64  = list(zip(['64neurons'] *64,  delta64))
del128 = list(zip(['128neurons']*128, delta128))
del256 = list(zip(['256neurons']*256, delta256))
del384 = list(zip(['384neurons']*384, delta384))
del512 = list(zip(['512neurons']*512, delta512))
del640 = list(zip(['640neurons']*640, delta640))
del768 = list(zip(['768neurons']*768, delta768))

df1 = pd.DataFrame(del64,  columns=['neuron size', 'vals'])
df2 = pd.DataFrame(del128, columns=['neuron size', 'vals'])
df3 = pd.DataFrame(del256, columns=['neuron size', 'vals'])
df4 = pd.DataFrame(del384, columns=['neuron size', 'vals'])
df5 = pd.DataFrame(del512, columns=['neuron size', 'vals'])
df6 = pd.DataFrame(del640, columns=['neuron size', 'vals'])
df7 = pd.DataFrame(del768, columns=['neuron size', 'vals'])

df = df1.append([df2, df3, df4, df5, df6, df7])

fig, ax = plt.subplots(figsize=(15, 10))

ax = sns.violinplot(x='neuron size', y='vals', data=df)
ax.grid()
ax.set_ylim([-0.2, 1.2])

# legend setting
ax.set_title('Adadelta selectivities batch_size=1')

plt.show()
```
